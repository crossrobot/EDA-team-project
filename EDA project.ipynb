### 2-1. 경쟁사 제품 리뷰 데이터 전처리

```python
# item 종류에 있는 각종 특수문자와 의미없는 문자 제거하기 : ': 세미콜론 기준 왼쪽에 있는 글자는 다 없애자.  
def modify1(item):
    x=re.sub(r'[^.]*:([^ ]*)', '', str(item))     ## 이거 알아내는데 5시간 걸렸다.. 실환가?
    return(x)   

review4rivals['item'] = review4rivals['item'].apply(modify1)

def modify2(item):                    # 특수문자 제거 
    x=re.sub(r'[^\s\w]','',str(item))               
    return(x)
review4rivals['item'] = review4rivals['item'].apply(modify2)

def modify3(item):                   #락토핏 앞에 거슬렸던 2, 9, 3과 같은 숫자들이 사라진다.  그러나 완벽하게 사라진 것은 아니다. 
    x=re.sub('\d\d*.?(락토핏)','락토핏',str(item))
    return(x)
review4rivals['item'] = review4rivals['item'].apply(modify3)

def modify4(item):
    x=re.sub('\d\d?\d?[가-힣]','',str(item))     # 몇통, 몇포와 같은 수량정보를 제거한다.
    return(x)
```

```python
print(review4rivals['item'].head(100))
print(review4rivals.item.tail(1000))
```

```python
def text_modify1(text):
    x=re.sub(r'[^가-힣|\d|\s]', '', str(text))     ## 사용자 리뷰에 있는 특수문자와 자, 모음 제거. 숫자는 살아있음. 
    return(x)   

review4rivals['text'] = review4rivals['text'].apply(text_modify1)
```
